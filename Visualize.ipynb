```python
df_mesh = pd.read_csv('mesh-16.txt', sep=' ')
df_tor = pd.read_csv('torus-16.txt', sep=' ')
```

```python
def set_plot_parameters (indx, title, fsize, data_x_m, data_y_m, col_m, data_x_t, data_y_t, col_t):
    ax[indx].plot(data_x_m, data_y_m, col_m, label='Mesh')
    ax[indx].plot(data_x_t, data_y_t, col_t, label='Torus')
    ax[indx].invert_xaxis()
    ax[indx].grid()
    ax[indx].set_title(title, {'fontsize': fsize})
    ax[indx].legend(fontsize = 13)
```
